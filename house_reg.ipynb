This plot makes it clear that most of the houses have 3 bedrooms

```python
x=df.drop(['id','price','zipcode','date','yr_built'],axis=1) #id has no importance, price is our target variable
                                                  #zipcode is dropped coz we have lat and long, so to avoid redundancy
                                                  #date is dropped just for simplicity 
y=df['price']
```

```python
x.head()
```

```python
#for normalizing 
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler() 
scaled_values = scaler.fit_transform(x) 
x.loc[:,:] = scaled_values
```

```python
x.head() #normalized values
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.40,random_state=42)
x_val,x_test,y_val,y_test=train_test_split(x_test,y_test,test_size=0.50,random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
print(x_val.shape)
print(y_val.shape)
```

```python
#fitting the model on training data
from sklearn.linear_model import LinearRegression
le=LinearRegression()
le.fit(x_train,y_train)
```

```python
#checking the score on validation data
le.score(x_val,y_val)
```

```python
#predicting the data on test data
y_pred=le.predict(x_test)
y_pred
```
